```python
# importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# loading the data
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/medical_students_dataset.csv")
```

```python
df.isnull()
df.isnull().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
# handling null in age
df['Age'].unique()
# print("Mean : ",df['Age'].mean())
# print("Mode : ",df['Age'].mode())
# print("Median : ",df['Age'].median())
df['Age'] = df['Age'].fillna(df['Age'].median()) # as median and mean have the same value -- 26.0
df['Age'] = df['Age'].astype(int)
df['Age'].isnull().sum()
```

```python
# handling null in Height
df['Height'].unique()
sorted(df['Age'].unique()) # no ouliers
print("Mean : ",df['Height'].mean())
# # print("Mode : ",df['Height'].mode())
print("Median : ",df['Height'].median())
df['Height'] = df['Height'].fillna(df['Height'].mean()) # as median and mean have the same value -- 174.9

df['Height'].isnull().sum()
```

```python
# handling null in Weight
df['Weight'].unique()
print("Mean : ",df['Weight'].mean())
# print("Mode : ",df['Weight'].mode())
print("Median : ",df['Weight'].median())
df['Weight'] = df['Weight'].fillna(df['Weight'].median()) # as median and mean have the same value -- 6

df['Weight'].isnull().sum()
```

```python
# handling null in BMI
df['BMI'].unique()
print("Mean : ",df['BMI'].mean())
# print("Mode : ",df['BMI'].mode())
print("Median : ",df['BMI'].median())
df['BMI'] = df['BMI'].fillna(df['BMI'].mean()) # as median and mean have the same value

df['BMI'].isnull().sum()
```

```python
# handling null in Temperature
df['Temperature'].unique()
print("Mean : ",df['Temperature'].mean())
# print("Mode : ",df['Temperature'].mode())
print("Median : ",df['Temperature'].median())
df['Temperature'] = df['Temperature'].fillna(df['Temperature'].mean()) # as median and mean have the same value
df['Temperature'].isnull().sum()
```

```python
# handling null in Heart Rate
df['Heart Rate'].unique()
print("Mean : ",df['Heart Rate'].mean())
print("Median : ",df['Heart Rate'].median())
df['Heart Rate'] = df['Heart Rate'].fillna(df['Heart Rate'].median()) # as mean giving decimals and no unique is float-- better to go with median and convert to int
df['Heart Rate'] = df['Heart Rate'].astype(int)
df['Heart Rate'].isnull().sum()
```

```python
# handling null in Blood Pressure
df['Blood Pressure'].unique()
print("Mean : ",df['Blood Pressure'].mean())
# print("Mode : ",df['Blood Pressure'].mode())
print("Median : ",df['Blood Pressure'].median())
df['Blood Pressure'] = df['Blood Pressure'].fillna(df['Blood Pressure'].median()) # as mean giving decimals and no unique is float-- better to go with median and convert to int
df['Blood Pressure'] = df['Blood Pressure'].astype(int)
df['Blood Pressure'].isnull().sum()
```

```python
# handling null in Cholesterol
df['Cholesterol'].unique()
print("Mean : ",df['Cholesterol'].mean())
print("Median : ",df['Cholesterol'].median())
df['Cholesterol'] = df['Cholesterol'].fillna(df['Cholesterol'].median()) # as mean giving decimals and no unique is float-- better to go with median and convert to int
df['Cholesterol'] = df['Cholesterol'].astype(int)
df['Cholesterol'].isnull().sum()
```

```python
# handling null in Diabetes
df['Diabetes'].unique()
print("Mode : ",df['Diabetes'].mode())

df['Diabetes'] = df['Diabetes'].fillna(df['Diabetes'].mode()[0]) # 1st value of mode

df['Diabetes'].isnull().sum()
```

```python
# handling null in Smoking
df['Smoking'].unique()
print("Mode : ",df['Smoking'].mode())

df['Smoking'] = df['Smoking'].fillna(df['Smoking'].mode()[0]) # 1st value of mode

df['Smoking'].isnull().sum()
```

```python
# handling null in Blood Type
df['Blood Type'].unique()
df['Blood Type'] = df['Blood Type'].fillna(df['Blood Type'].mode()[0])
df['Blood Type'].unique()
```

```python
# handling null in Gender
df['Gender'].unique()
# print("Mode : ",df['Gender'].mode())

df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0]) # 1st value of mode

df['Gender'].isnull().sum()
```

```python
# handling StudentID
df.dropna(inplace = True)
```
